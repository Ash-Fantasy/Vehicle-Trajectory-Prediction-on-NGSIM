```python
import csv
import pandas as pd
import numpy as np
root='/home/mt/learn/NGSIM数据处理/csv/80/5:15-5:30'
f1=pd.read_csv(root+"/Doubel_label.csv")
f2=pd.read_csv('/home/mt/learn/NGSIM数据处理/csv/80/5:00-5:15/Double_label1.csv')
f3=pd.read_csv('/home/mt/learn/NGSIM数据处理/csv/80/5:15-5:30/Double_label1.csv')
f4=pd.read_csv('/home/mt/learn/NGSIM数据处理/csv/80/5:00-5:15/Double_train1.csv')
f5=pd.read_csv('/home/mt/learn/NGSIM数据处理/csv/80/5:15-5:30/Double_train1.csv')
f1
```

```python
fr=f1[f1.label.isin([2])]
fl=f1[f1.label.isin([0,1])]
fl=fl.reset_index(drop=True)
fl
```

```python
fr_index=fr.index
fr_x=fr[['Vehicle_ID','Frame_ID','Local_X']].values
fr_index=np.reshape(fr_index,[-1,1])
fr_arr=np.hstack((fr_index,fr_x))
fr_arr=np.reshape(fr_arr,[-1,40,4])
len(fr_arr)*40
```

```python
shapes=np.shape(fr_arr)
rows=[]
for i in  range(shapes[0]):
    if fr_arr[i][39][3]-fr_arr[i][0][3]>=2:
        rows.extend(int(fr_arr[i][a][0]) for a in range(40))      
len(rows)    
```

```python
fr_new=fr.loc[rows]
fr_new=fr_new.reset_index(drop=True)
fr_new
```

```python
f_new=pd.concat([fl,fr_new])
f_new
```

```python
def normalize(train):
  train_norm = train.apply(lambda x: (x - np.mean(x)) / (np.max(x) - np.min(x)))
  return train_norm
```

```python
f_new.to_csv(root+"/Double_label1.csv",index=False)
df_combine=normalize(f_new)
df_combine.to_csv(root+"/Double_train1.csv",index=False)
```

```python
f_label=pd.concat([f2,f3[f3.label.isin([2])]])
f_train=normalize(f_label)
f_label.to_csv('/home/mt/learn/NGSIM数据处理/csv/合并数据/Double_label1.csv',index=False)
f_train.to_csv('/home/mt/learn/NGSIM数据处理/csv/合并数据/Double_train1.csv',index=False)
```
