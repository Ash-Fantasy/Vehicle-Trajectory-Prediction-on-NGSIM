```python
import matplotlib.pyplot as plt
import csv
import pandas as pd
import numpy as np
f=pd.read_csv("/home/mt/learn/NGSIM数据处理/csv/80/4:00-4:15/change_lane_id.csv")
local=pd.read_csv("/home/mt/learn/NGSIM数据处理/csv/80/4:00-4:15/one_change_local.csv")
vehicle_id=[]
a=list(f['Vehicle_ID'])
for i in a:
    if i not in vehicle_id:
        vehicle_id.append(i)
x_y=f[['Local_X','Local_Y']]
len(vehicle_id)
```

```python
change_x=list(local.Local_X)
change_y=list(local.Local_Y)
change_x
```

```python
x_list=[]
y_list=[]
for i in vehicle_id:
    x=f[f.Vehicle_ID.isin([i])]
    x_list.append(list(x['Local_X']))
    y_list.append(list(x['Local_Y']))
```
