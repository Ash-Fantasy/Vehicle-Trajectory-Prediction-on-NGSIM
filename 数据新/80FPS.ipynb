```python
import csv
import pandas as pd
import numpy as np
root='/home/mt/learn/NGSIM数据处理/csv/80/4:00-4:15'
f1=pd.read_csv(root+"/Double_label1.csv")
f2=pd.read_csv('/home/mt/learn/NGSIM数据处理/csv/80/算法数据/4:00-4:15.csv')
f2
```

```python
vf=f1[['Vehicle_ID','Frame_ID','label']].values
vf=np.reshape(vf,[-1,40,3])
vf
```

```python
start_frame=[]
mid_frame=[]
end_frame=[]
id_list=[]
all_frame=[]
label_list=[]
for i in range(len(vf)):
    id_list.append(vf[i][0][0])
    start_frame.append(vf[i][0][1])
    mid_frame.append(vf[i][-1][1])
    end_frame.append(vf[i][-1][1]+40)
    label_list.append(vf[i][-1][2])
for i in range(len(start_frame)):
    all_frame.extend(start_frame[i]+a for a in range(80))
all_frame=np.reshape(all_frame,[-1,80])
label_list
```

```python
f3=f2[f2.Vehicle_ID.isin(id_list)]
f3
```

```python
index_list=[]
drop_list=[]
total_list=[]
for i in range(len(vf)):
    a1=f3[f3.Vehicle_ID.isin([id_list[i]])]
    b1=a1[a1.Frame_ID.isin(all_frame[i])]
    c1=list(b1.index)
    index_list.extend(c1[0]+a for a in range(80))
    if len(b1)!=80:
        drop_list.append(c1)
        total_list.extend(c1[0]+a for a in range(80))
drop_list
```

```python
index_list
```

```python
label=[]
for i in range(len(vf)):
    for a in range(80):
        label.append(label_list[i])
len(label)        
```

```python
lane_local= pd.DataFrame(data=label,columns = ["label"]) 
lane_local
```

```python
new_list=[]
# index_list1=sum(index_list,[])
drop_list1=sum(drop_list,[])
for i in total_list:
    if i not in drop_list1:
        new_list.append(i)
new_list   
```

```python
len(new_list)
```

```python
pd1=f2.loc[index_list]
pd1.loc[new_list]=''
pd1=pd1.reset_index(drop=True)
pd80=pd.concat([pd1,lane_local],axis=1)
```
