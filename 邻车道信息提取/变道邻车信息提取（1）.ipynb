```python
import numpy as np
import pandas as pd
import random

f=pd.read_csv("/home/mt/learn/NGSIM数据处理/csv/80/4:00-4:15/CL_train41.csv")
f=f[['Vehicle_ID','Frame_ID','Local_X','Local_Y','Preceding','Following','Lane_ID']]
f1=pd.read_csv("/media/mt/Seagate Backup Plus Drive/csv/80/vehicle-trajectory-data/0400pm-0415pm/trajectories-0400-0415.csv")

f
```

```python
def repeat(re_list):
    a=[]
    for i in re_list:
        if i not in a:
            a.append(i)
    return a
```

```python
data_num=int(len(f)/41)
data_num
```

```python
f_list=f.values
f_list=np.reshape(f_list,[-1,41,7])
np.shape(f_list)
```

```python
veh_num=max(f1.Vehicle_ID.values)
zero_list=[]
for i in range(len(f_list)):
    for a in range(41):
        if f_list[i][a,4]==0 or f_list[i][a,5]==0 or f_list[i][0,4]!=f_list[i][39,4] or f_list[i][0,5]!=f_list[i][39,5] or f_list[i][a,4]>veh_num or f_list[i][a,5]>veh_num:
            zero_list.append(i)
zero_list=repeat(re_list=zero_list) 
zero_list
```

```python
f_list=np.delete(f_list,zero_list, axis=0)
len(f_list)
```

```python
f_list=f_list.reshape([-1,7])
f_new=pd.DataFrame(f_list,columns=['Vehicle_ID','Frame_ID','Local_X','Local_Y','Preceding','Following','Lane_ID'])
f_new.to_csv("/home/mt/learn/NGSIM数据处理/csv/80/变道邻车信息/new_RL_label.csv",index=0)
```
